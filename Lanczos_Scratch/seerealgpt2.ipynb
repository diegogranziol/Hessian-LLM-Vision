```python
import torch
import matplotlib.pyplot as plt

# Path to the checkpoint file
checkpoint_path = 'model_trained.pt.ckpt'

# Load the checkpoint into a variable
checkpoint = torch.load('model_trained.pt.ckpt',map_location=torch.device('cpu'))

# Assuming 'eigvals' and 'gammas' are keys in the checkpoint dictionary
eigvals = checkpoint['eigvals'].numpy()
gammas = checkpoint['gammas'].numpy()

# Plotting using stem
plt.figure()
plt.stem(eigvals, gammas, use_line_collection=True)
plt.xlabel('Eigenvalues')
plt.ylabel('Gammas')
plt.yscale('log')

plt.title('Stem Plot of Eigenvalues and Gammas')
plt.grid(True)
plt.show()
```
